```python
from sklearn.feature_selection import SelectPercentile, f_classif

selector=SelectPercentile(f_classif, percentile=10)
X_newtrain = selector.fit_transform(X_train,Y_train)
X_newtrain.shape
```

```python
selector=SelectPercentile(f_classif, percentile=10)
X_newtest = selector.fit_transform(X_test,Y_test)
X_newtest.shape
```

```python
true=selector.get_support()
X_trainnewcolindex=[i for i, x in enumerate(true) if x]
X_trainnewcolname=list( colNames[i] for i in X_trainnewcolindex )
X_trainnewcolname
```

```python
#with a naive bayes
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC


clf = MultinomialNB().fit(X_newtrain, Y_train)
print("naive bayes accuracy is")

#step 4: evaluate classifier

print(clf.score(X_newtest, Y_test))
```
